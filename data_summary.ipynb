```python
import pandas as pd
```

```python
df=pd.read_csv('../Data/2D/2d.csv')
```

```python
df
```

```python
df['image_type']=df['image_type'].replace("x-ray","X-ray")
df['image_type']=df['image_type'].replace("xRay","X-Ray")
df['image_type']=df['image_type'].replace("Xray","X-Ray")
df['image_type']=df['image_type'].replace("X-Ray","X-ray")
df['image_type']=df['image_type'].str.replace(' ', '')
df['region']=df['region'].str.replace('chest', 'Chest')
```

```python
unique_class=df['dataset_name'].unique()
```

```python
def summarize_data(df):
    summary = {
        'X-ray': {},
        'CT': {},
        'MRI': {},

    }
    
    # Filter the dataframe by image type and then perform calculations
    for image_type in summary.keys():
        filtered_df = df[df['image_type'].str.contains(image_type, case=False)]
        
        # Group by 'region' to calculate the statistics for each anatomy
        for region, group in filtered_df.groupby('region'):
            number_of_classes = group['dataset_name'].nunique()
            total_images = len(group)
            average_class_size = int(total_images / number_of_classes) if number_of_classes > 0 else 0
            
            # Populate the summary dictionary
            summary[image_type][region] = {
                '# Classes': number_of_classes,
                'Average class size': average_class_size,
                'Total # image': total_images
            }
    
    return summary

# Calculate the summary statistics
data_summary = summarize_data(df)

# To display or save the summary, you can use pandas DataFrame again or any other preferred format
summary_df = pd.DataFrame.from_dict({(i,j): data_summary[i][j] 
                                     for i in data_summary.keys() 
                                     for j in data_summary[i].keys()},
                                     orient='index')

print(summary_df)
```

```python
summary_df.to_csv('../summary.csv')
```
